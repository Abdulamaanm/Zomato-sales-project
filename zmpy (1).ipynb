# convert  the data type of column  rate

```python
import pandas as pd

# Load the data
zdata = pd.read_csv("Zomato_data.csv")

# Define the rate handling function
def handlerate(value):
    try:
        value = str(value).split("/")[0].strip()  # Split by '/' and take the first part
        return float(value)  # Convert to float
    except ValueError:
        return None  

# Apply the function to the 'rate' column
zdata["rate"] = zdata["rate"].apply(handlerate)

# Optionally, convert to integer if you want to remove fractional values
zdata["rate"] = zdata["rate"].fillna(0).astype(int)  # Replace NaN with 0 or choose another strategy

zdata
```
